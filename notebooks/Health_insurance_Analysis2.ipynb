```python
import pandas as pd

COLUMN_RENAME_MAP = {
    # IDs
    "BENEFICIARY_ID": "beneficiary_id",
    "DESYNPUF_ID": "beneficiary_id",
    "Beneficiory_ID": "beneficiary_id",  # 2010 labelled (typo)

    # Dates
    "BENE_BIRTH_DT": "birth_date",
    "DOB": "birth_date",
    "BENE_DEATH_DT": "death_date",
    "Date of Death": "death_date",

    # Demographics
    "BENE_SEX_IDENT_CD": "sex_code",
    "SEX": "sex_code",
    "BENE_RACE_CD": "race_code",
    "RACE": "race_code",

    # ESRD
    "BENE_ESRD_IND": "esrd_indicator",
    "End-Stage Renal Disease indicator": "esrd_indicator",

    # Geography
    "SP_STATE_CODE": "state_code",
    "STATE": "state_code",
    "BENE_COUNTY_CD": "county_code",
    "COUNTY": "county_code",

    # Coverage
    "BENE_HI_CVRAGE_TOT_MONS": "part_a_coverage_months",
    "Part A (Hospital insurance)": "part_a_coverage_months",
    "BENE_SMI_CVRAGE_TOT_MONS": "part_b_coverage_months",
    "Part B (Medical insurance)": "part_b_coverage_months",
    "BENE_HMO_CVRAGE_TOT_MONS": "hmo_coverage_months",
    "HMO": "hmo_coverage_months",
    "PLAN_CVRG_MOS_NUM": "total_coverage_months",
    "Total months of any plan coverage": "total_coverage_months",

    # Chronic flags
    "SP_ALZHDMTA": "chronic_alzheimer",
    "SP_CHF": "chronic_heart_failure",
    "SP_CHRNKIDN": "chronic_kidney_disease",
    "SP_CNCR": "chronic_cancer",
    "SP_COPD": "chronic_copd",
    "SP_DEPRESSN": "chronic_depression",
    "SP_DIABETES": "chronic_diabetes",
    "SP_ISCHMCHT": "chronic_ischemic_heart_disease",
    "SP_OSTEOPRS": "chronic_osteoporosis",
    "SP_RA_OA": "chronic_rheumatoid_arthritis_osteoarthritis",
    "SP_STRKETIA": "chronic_stroke_tia",

    # Payments
    "MEDREIMB_IP": "medicare_payment_inpatient",
    "Medicare PaidAmount_IP": "medicare_payment_inpatient",
    "BENRES_IP": "beneficiary_payment_inpatient",
    "PPPYMT_IP": "third_party_payment_inpatient",

    "MEDREIMB_OP": "medicare_payment_outpatient",
    "BENRES_OP": "beneficiary_payment_outpatient",
    "PPPYMT_OP": "third_party_payment_outpatient",

    "MEDREIMB_CAR": "medicare_payment_carrier",
    "BENRES_CAR": "beneficiary_payment_carrier",
    "PPPYMT_CAR": "third_party_payment_carrier",
}

# Only raw chronic columns need conversion (labelled file already has chronic_* booleans)
CHRONIC_RAW_COLS = [
    "SP_ALZHDMTA", "SP_CHF", "SP_CHRNKIDN", "SP_CNCR", "SP_COPD",
    "SP_DEPRESSN", "SP_DIABETES", "SP_ISCHMCHT", "SP_OSTEOPRS",
    "SP_RA_OA", "SP_STRKETIA"
]


def prepare_beneficiary_year(
    df: pd.DataFrame,
    year: int,
    add_features: bool = True,
    date_format: str = "%Y%m%d",
) -> pd.DataFrame:
    """
    Clean + standardise a DE-SynPUF beneficiary dataframe for a given year.
    Handles both CMS raw uppercase columns and 2010 labelled/human-readable columns.
    """
    df = df.copy()

    # 0) Normalise column whitespace (critical for ' Part A ...')
    df.columns = df.columns.astype(str).str.strip()

    # 1) Convert chronic flags if raw cols exist (SP_* only)
    present_chronic = [c for c in CHRONIC_RAW_COLS if c in df.columns]
    for c in present_chronic:
        df[c] = pd.to_numeric(df[c], errors="coerce").eq(1).astype("boolean")

    # 2) ESRD: handle both raw and labelled variants BEFORE rename
    if "BENE_ESRD_IND" in df.columns:
        s = df["BENE_ESRD_IND"]
        if s.dtype == "object":
            df["BENE_ESRD_IND"] = s.replace({"Y": 1, "N": 0}).fillna(0).astype(int)
        else:
            df["BENE_ESRD_IND"] = pd.to_numeric(s, errors="coerce").fillna(0).astype(int)

    if "End-Stage Renal Disease indicator" in df.columns:
        s = df["End-Stage Renal Disease indicator"]
        if s.dtype == "object":
            df["End-Stage Renal Disease indicator"] = s.replace({"Y": 1, "N": 0}).fillna(0).astype(int)
        else:
            df["End-Stage Renal Disease indicator"] = pd.to_numeric(s, errors="coerce").fillna(0).astype(int)

    # 3) Rename columns to canonical names
    rename_map_present = {k: v for k, v in COLUMN_RENAME_MAP.items() if k in df.columns}
    df = df.rename(columns=rename_map_present)

    # 4) Parse dates AFTER rename (so both raw and labelled become birth_date/death_date)
    if "birth_date" in df.columns:
        # if it's numeric yyyymmdd, parse with format; otherwise let pandas infer
        if pd.api.types.is_numeric_dtype(df["birth_date"]):
            df["birth_date"] = pd.to_datetime(df["birth_date"].astype("Int64").astype(str), format=date_format, errors="coerce")
        else:
            df["birth_date"] = pd.to_datetime(df["birth_date"], errors="coerce")

    if "death_date" in df.columns:
        if pd.api.types.is_numeric_dtype(df["death_date"]):
            df["death_date"] = pd.to_datetime(df["death_date"].astype("Int64").astype(str), format=date_format, errors="coerce")
        else:
            df["death_date"] = pd.to_datetime(df["death_date"], errors="coerce")

    # 5) Add derived features
    if add_features:
        df["year"] = year

        # Use age_at_year instead of age (dimension-friendly)
        if "birth_date" in df.columns:
            df["age_at_year"] = (year - df["birth_date"].dt.year).astype("Int64")

        # Alive flag at end of that year
        if "death_date" in df.columns:
            df["is_alive_flag"] = df["death_date"].isna() | (df["death_date"].dt.year > year)
        else:
            df["is_alive_flag"] = pd.Series([True] * len(df), dtype="boolean")

        chronic_cols = [c for c in df.columns if c.startswith("chronic_")]
        if chronic_cols and "chronic_condition_count" not in df.columns:
            df["chronic_condition_count"] = df[chronic_cols].sum(axis=1)

        # Totals (I still think these should move to FACT later, but keeping your current design)
        if df.filter(like="medicare_payment_").shape[1] > 0:
            df["total_medicare_payment"] = df.filter(like="medicare_payment_").sum(axis=1)
        if df.filter(like="beneficiary_payment_").shape[1] > 0:
            df["total_beneficiary_payment"] = df.filter(like="beneficiary_payment_").sum(axis=1)
        if df.filter(like="third_party_payment_").shape[1] > 0:
            df["total_third_party_payment"] = df.filter(like="third_party_payment_").sum(axis=1)

        if {"total_medicare_payment", "total_beneficiary_payment", "total_third_party_payment"}.issubset(df.columns):
            df["total_payment"] = (
                df["total_medicare_payment"]
                + df["total_beneficiary_payment"]
                + df["total_third_party_payment"]
            )

    return df
```

```python
import pandas as pd
from typing import Optional

# ---------- Configuration ----------
INPATIENT_RENAME_MAP = {
    "DESYNPUF_ID": "beneficiary_id",
    "CLM_ID": "claim_id",
    "PRVDR_NUM": "provider_id",

    "CLM_FROM_DT": "claim_start_date",
    "CLM_THRU_DT": "claim_end_date",
    "CLM_ADMSN_DT": "admission_date",
    "NCH_BENE_DSCHRG_DT": "discharge_date",

    "CLM_PMT_AMT": "medicare_payment_amount",
    "NCH_PRMRY_PYR_CLM_PD_AMT": "primary_payer_paid_amount",

    "CLM_UTLZTN_DAY_CNT": "utilization_days",
    "CLM_DRG_CD": "drg_code",

    "ADMTNG_ICD9_DGNS_CD": "primary_diagnosis",
}

DATE_COLS = ["claim_start_date", "claim_end_date", "admission_date", "discharge_date"]

NUMERIC_COLS = [
    "medicare_payment_amount",
    "primary_payer_paid_amount",
    "utilization_days",
]

REQUIRED_COLS = ["beneficiary_id", "claim_id", "admission_date", "discharge_date"]


def _parse_yyyymmdd(series: pd.Series, fmt: str = "%Y%m%d") -> pd.Series:
    """Parse yyyymmdd dates stored as int/float/str."""
    s = series.copy()
    if pd.api.types.is_numeric_dtype(s):
        s = s.astype("Int64").astype(str)
    else:
        s = s.astype(str).str.strip()
    return pd.to_datetime(s, format=fmt, errors="coerce")


def clean_inpatient_claims(
    df: pd.DataFrame,
    year_hint: Optional[int] = None,
    date_format: str = "%Y%m%d",
    keep_optional_code_columns: bool = True,
) -> pd.DataFrame:
    """
    Clean + standardise DE-SynPUF inpatient claims at claim-grain (1 row = 1 claim).
    Produces a staging dataframe suitable for loading to SQL as stg_inpatient_claims.
    """
    df = df.copy()

    # 0) Normalize headers
    df.columns = df.columns.astype(str).str.strip()

    # 1) Rename columns (only those present)
    rename_present = {k: v for k, v in INPATIENT_RENAME_MAP.items() if k in df.columns}
    df = df.rename(columns=rename_present)

    # 2) Parse dates
    for c in DATE_COLS:
        if c in df.columns:
            df[c] = _parse_yyyymmdd(df[c], fmt=date_format)

    # 3) Clean numerics
    for c in NUMERIC_COLS:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    # 4) Strip ids/codes
    for c in ["beneficiary_id", "claim_id", "provider_id", "drg_code", "primary_diagnosis"]:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip().replace({"": pd.NA})

    # 5) Drop bad rows missing required fields
    present_required = [c for c in REQUIRED_COLS if c in df.columns]
    if present_required:
        df = df.dropna(subset=present_required).copy()

    # 6) Derive length_of_stay
    df["length_of_stay"] = (df["discharge_date"] - df["admission_date"]).dt.days
    df.loc[df["length_of_stay"] < 0, "length_of_stay"] = pd.NA

    # 7) Derive year (use admission_date year)
    df["year"] = df["admission_date"].dt.year.astype("Int64")
    if df["year"].isna().all() and year_hint is not None:
        df["year"] = pd.Series([year_hint] * len(df), dtype="Int64")

    # 8) Claim grain check: remove duplicates by claim_id
    if "claim_id" in df.columns:
        dup_mask = df["claim_id"].duplicated(keep="first")
        df["claim_id_duplicated_flag"] = dup_mask
        df = df[~dup_mask].copy()

    # 9) Optional: clean wide code columns (keep them, but strip whitespace)
    if keep_optional_code_columns:
        code_prefixes = ("ICD9_DGNS_CD_", "ICD9_PRCDR_CD_", "HCPCS_CD_")
        for c in df.columns:
            if c.startswith(code_prefixes):
                df[c] = df[c].astype(str).str.strip().replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})

    # 10) Order core columns first
    core_order = [
        "claim_id",
        "beneficiary_id",
        "provider_id",
        "admission_date",
        "discharge_date",
        "claim_start_date",
        "claim_end_date",
        "year",
        "length_of_stay",
        "utilization_days",
        "drg_code",
        "primary_diagnosis",
        "medicare_payment_amount",
        "primary_payer_paid_amount",
        "claim_id_duplicated_flag",
    ]
    existing_core = [c for c in core_order if c in df.columns]
    remaining = [c for c in df.columns if c not in existing_core]
    return df[existing_core + remaining]
```

```python
import pandas as pd
from typing import Optional

OUTPATIENT_RENAME_MAP = {
    "DESYNPUF_ID": "beneficiary_id",
    "CLM_ID": "claim_id",
    "PRVDR_NUM": "provider_id",

    "CLM_FROM_DT": "claim_start_date",
    "CLM_THRU_DT": "claim_end_date",

    "CLM_PMT_AMT": "medicare_payment_amount",
    "NCH_PRMRY_PYR_CLM_PD_AMT": "primary_payer_paid_amount",

    "NCH_BENE_BLOOD_DDCTBL_LBLTY_AM": "blood_deductible_liability_amount",
    "NCH_BENE_PTB_DDCTBL_AMT": "part_b_deductible_amount",
    "NCH_BENE_PTB_COINSRNC_AMT": "part_b_coinsurance_amount",

    "ADMTNG_ICD9_DGNS_CD": "primary_diagnosis",
}

OUTPATIENT_DATE_COLS = ["claim_start_date", "claim_end_date"]

OUTPATIENT_NUMERIC_COLS = [
    "medicare_payment_amount",
    "primary_payer_paid_amount",
    "blood_deductible_liability_amount",
    "part_b_deductible_amount",
    "part_b_coinsurance_amount",
]

OUTPATIENT_REQUIRED_COLS = ["beneficiary_id", "claim_id", "claim_start_date"]


def _parse_yyyymmdd(series: pd.Series, fmt: str = "%Y%m%d") -> pd.Series:
    s = series.copy()
    if pd.api.types.is_numeric_dtype(s):
        s = s.astype("Int64").astype(str)
    else:
        s = s.astype(str).str.strip()
    return pd.to_datetime(s, format=fmt, errors="coerce")


def clean_outpatient_claims(
    df: pd.DataFrame,
    year_hint: Optional[int] = None,
    date_format: str = "%Y%m%d",
    drop_all_null_hcpcs: bool = False,
) -> pd.DataFrame:
    """
    Clean + standardise DE-SynPUF outpatient claims at claim-grain (1 row = 1 claim).
    Produces stg_outpatient_claims_clean.
    """
    df = df.copy()

    # 0) Normalize headers
    df.columns = df.columns.astype(str).str.strip()

    # 1) Rename (present only)
    rename_present = {k: v for k, v in OUTPATIENT_RENAME_MAP.items() if k in df.columns}
    df = df.rename(columns=rename_present)

    # 2) Parse dates
    for c in OUTPATIENT_DATE_COLS:
        if c in df.columns:
            df[c] = _parse_yyyymmdd(df[c], fmt=date_format)

    # 3) Clean numerics
    for c in OUTPATIENT_NUMERIC_COLS:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    # 4) Strip IDs/codes
    for c in ["beneficiary_id", "claim_id", "provider_id", "primary_diagnosis"]:
        if c in df.columns:
            df[c] = df[c].astype(str).str.strip().replace({"": pd.NA})

    # 5) Drop bad rows missing required fields
    present_required = [c for c in OUTPATIENT_REQUIRED_COLS if c in df.columns]
    if present_required:
        df = df.dropna(subset=present_required).copy()

    # 6) Year (use claim_start_date year)
    if "claim_start_date" in df.columns:
        df["year"] = df["claim_start_date"].dt.year.astype("Int64")
    elif year_hint is not None:
        df["year"] = pd.Series([year_hint] * len(df), dtype="Int64")

    # 7) Claim grain check: remove duplicates by claim_id
    if "claim_id" in df.columns:
        dup_mask = df["claim_id"].duplicated(keep="first")
        df["claim_id_duplicated_flag"] = dup_mask
        df = df[~dup_mask].copy()
    else:
        df["claim_id_duplicated_flag"] = False

    # 8) Optional: clean code columns (ICD/HCPCS) whitespace
    code_prefixes = ("ICD9_DGNS_CD_", "ICD9_PRCDR_CD_", "HCPCS_CD_")
    for c in df.columns:
        if c.startswith(code_prefixes):
            df[c] = df[c].astype(str).str.strip().replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})

    # 9) Optional: drop HCPCS columns that are ALL null (only do this after checking!)
    if drop_all_null_hcpcs:
        hcpcs_cols = [c for c in df.columns if c.startswith("HCPCS_CD_")]
        all_null = [c for c in hcpcs_cols if df[c].notna().sum() == 0]
        if all_null:
            df = df.drop(columns=all_null)

    # 10) Order core columns first
    core_order = [
        "claim_id",
        "beneficiary_id",
        "provider_id",
        "claim_start_date",
        "claim_end_date",
        "year",
        "primary_diagnosis",
        "medicare_payment_amount",
        "primary_payer_paid_amount",
        "part_b_deductible_amount",
        "part_b_coinsurance_amount",
        "blood_deductible_liability_amount",
        "claim_id_duplicated_flag",
        "SEGMENT",  # keep if present, optional
        "AT_PHYSN_NPI",
        "OP_PHYSN_NPI",
        "OT_PHYSN_NPI",
    ]
    existing_core = [c for c in core_order if c in df.columns]
    remaining = [c for c in df.columns if c not in existing_core]
    return df[existing_core + remaining]
```
